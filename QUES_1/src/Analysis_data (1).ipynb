```python
import pandas as pd

# Read an Excel file into a DataFrame
df = pd.read_csv('clean_data.csv')

# Print the DataFrame
print(df)
```

```python
correlation_matrix = df.corr()

print(correlation_matrix)
```

```python
# Convert frailty labels to numerical values
frailty_numeric = df['Frailty'].replace({'Y': 1, 'N': 0})

# Add the numeric frailty column to the DataFrame
df['Frailty_Numeric'] = frailty_numeric

# Calculate the correlation between frailty and other variables
correlation_with_frailty = df.corr()['Frailty_Numeric'].drop('Frailty_Numeric')

print(correlation_with_frailty)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(df['Age'], df['Frailty'], color='blue')
plt.title('Age vs. Frailty')
plt.xlabel('Age')
plt.ylabel('Frailty (0: No, 1: Yes)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.subplot(1, 2, 2)
plt.scatter(df['Age'], df['GripStrength'], color='green')
plt.title('Age vs. Grip strength')
plt.xlabel('Age')
plt.ylabel('Grip Strength')
plt.grid(True)
```
